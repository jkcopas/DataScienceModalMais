##3. Obtenção via Biblioteca

Como exemplo, vamos utilizar a função "download()" da biblioteca "yfinance" que nos possibilita extrair dados de açoes e de varios outros tipos de ativos 

```python
#Extração de um ativo PETR4

yf.download('PETR4.SA',start = '2022-02-01', end = '2022-02-18', period ='1d')
```

```python
lren = yf.download('LREN3.SA', start = '2017-01-01',end = '2021-07-20')
```

```python
lren.Close.plot()
```

```python
# importand omultiplos ativos para um dataframe

dados = yf.download("BBAS3.SA BPAC11.SA BBDC4.SA", start = '2019-01-01', end ='2021-12-31')
dados
```

```python
dados = yf.download("BBAS3.SA BPAC11.SA BBDC4.SA", start = '2019-01-01', end ='2021-12-31', group_by='tickers')
dados
```

```python
#historico maximo de dados de um ativo

vale = yf.Ticker("VALE3.SA")
vale_historico = vale.history(period='max', interval = '1wk')
vale_historico
```
